# Loading SEGY files from the Northern Lights well 31/5-7
Part of the data pack for the Northern Lights well 31/5-7 are SEGY files containing data from the Zero Offset Vertical Seismic Profile survey (ZOVSP) that was acquired. These can be opened using the package segyio. Examples of the data are shown below.

```python
import segyio
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#Defining file names for the raw data files
raw1='Northern_Lights/07.Borehole_Seismic/VSPZO_RAW_2020-01-17_1.SEGY'
raw2='Northern_Lights/07.Borehole_Seismic/VSPZO_RAW_2020-01-17_2.SEGY'
raw3='Northern_Lights/07.Borehole_Seismic/VSPZO_RAW_2020-01-17_3.SEGY'
raw4='Northern_Lights/07.Borehole_Seismic/VSPZO_RAW_2020-01-17_4.SEGY'

#Defining file names for the corridor stack data files
corridorstack1='Northern_Lights/07.Borehole_Seismic/VSPZO_COMPUTED_TIME_CS_2020-01-17_1.SEGY'
corridorstack2='Northern_Lights/07.Borehole_Seismic/VSPZO_COMPUTED_TIME_CS_2020-01-17_2.SEGY'
corridorstack3='Northern_Lights/07.Borehole_Seismic/VSPZO_COMPUTED_TIME_CS_2020-01-17_3.SEGY'
corridorstack4='Northern_Lights/07.Borehole_Seismic/VSPZO_COMPUTED_TIME_CS_2020-01-17_4.SEGY'

#Defining file names for a selection of the time gather data files
timegather1='Northern_Lights/07.Borehole_Seismic/VSPZO_COMPUTED_TIME_GATHER_2020-01-17_12.SEGY'
timegather2='Northern_Lights/07.Borehole_Seismic/VSPZO_COMPUTED_TIME_GATHER_2020-01-17_13.SEGY'
timegather3='Northern_Lights/07.Borehole_Seismic/VSPZO_COMPUTED_TIME_GATHER_2020-01-17_19.SEGY'
timegather4='Northern_Lights/07.Borehole_Seismic/VSPZO_COMPUTED_TIME_GATHER_2020-01-17_23.SEGY'
```

```python
#Open SEGY files for the raw data
with segyio.open(raw1, strict=False) as s:
    rawdata1 = np.stack(t.astype(np.float64) for t in s.trace)

with segyio.open(raw2, strict=False) as s:
    rawdata2 = np.stack(t.astype(np.float64) for t in s.trace)
    
with segyio.open(raw3, strict=False) as s:
    rawdata3 = np.stack(t.astype(np.float64) for t in s.trace)

with segyio.open(raw4, strict=False) as s:
    rawdata4 = np.stack(t.astype(np.float64) for t in s.trace)
```

```python
#Showing the loaded raw data    
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey=True, figsize=(12,8))
fig.suptitle('RAW DATA - Files: VSPZO_RAW_2020-01-17_n.SEGY')

ax1.imshow(rawdata1.T,cmap='seismic',vmin=-0.1, vmax=0.1,aspect='auto')
ax1.set_title('SURFACE REF. HYDROPHONE')
ax1.set_ylabel('Samples, depth Sampling Rate = 1 ms')
ax1.set_xlabel('Shot Number')

ax2.imshow(rawdata2.T,cmap='seismic',vmin=-0.1, vmax=0.1,aspect='auto')
ax2.set_title('DOWNHOLE Z RAW SHOTS')
ax2.set_xlabel('time (ms)')

ax3.imshow(rawdata3.T,cmap='seismic',vmin=-0.1, vmax=0.1,aspect='auto')
ax3.set_title('DOWNHOLE X RAW SHOTS')
ax3.set_xlabel('time (ms)')

ax4.imshow(rawdata4.T,cmap='seismic',vmin=-0.1, vmax=0.1,aspect='auto')
ax4.set_title('DOWNHOLE Y RAW SHOTS')
ax4.set_xlabel('time (ms)')

plt.show()
```

```python
#Open SEGY files for the corridor stack data
with segyio.open(corridorstack1, strict=False) as s:
    csdata1 = np.stack(t.astype(np.float64) for t in s.trace)

with segyio.open(corridorstack2, strict=False) as s:
    csdata2 = np.stack(t.astype(np.float64) for t in s.trace)
    
with segyio.open(corridorstack3, strict=False) as s:
    csdata3 = np.stack(t.astype(np.float64) for t in s.trace)

with segyio.open(corridorstack4, strict=False) as s:
    csdata4 = np.stack(t.astype(np.float64) for t in s.trace)
```

```python
#Showing the loaded corridor stack data 
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey=True, figsize=(12,8))
fig.suptitle('ZERO PHASE CORRIDOR STACK - Files: VSPZO_COMPUTED_TIME_CS_2020-01-17_n.SEGY')

ax1.imshow(csdata1.T,cmap='seismic',vmin=-5, vmax=5,aspect='auto')
ax1.set_title('5-90 HZ')
ax1.set_ylabel('Samples, depth Sampling Rate = 1 ms')
ax1.set_xlabel('trace number')

ax2.imshow(csdata2.T,cmap='seismic',vmin=-5, vmax=5,aspect='auto')
ax2.set_title('5-70 HZ')
ax2.set_xlabel('trace number')

ax3.imshow(csdata3.T,cmap='seismic',vmin=-5, vmax=5,aspect='auto')
ax3.set_title('5-50 HZ')
ax3.set_xlabel('trace number')

ax4.imshow(csdata4.T,cmap='seismic',vmin=-5, vmax=5,aspect='auto')
ax4.set_title('5-40 HZ')
ax4.set_xlabel('trace number')

plt.show()
```

```python
#Open SEGY files for the time gather data
with segyio.open(timegather1, strict=False) as s:
    tgdata1 = np.stack(t.astype(np.float64) for t in s.trace)

with segyio.open(timegather2, strict=False) as s:
    tgdata2 = np.stack(t.astype(np.float64) for t in s.trace)
    
with segyio.open(timegather3, strict=False) as s:
    tgdata3 = np.stack(t.astype(np.float64) for t in s.trace)

with segyio.open(timegather4, strict=False) as s:
    tgdata4 = np.stack(t.astype(np.float64) for t in s.trace)
```

```python
#Showing the loaded time gather data 
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey=True, figsize=(12,8))
fig.suptitle('SELETION OF COMPUTED TIME GATHER DATA - Files: VSPZO_COMPUTED_TIME_GATHER_2020-01-17_n.SEGY')

ax1.imshow(tgdata1.T,cmap='seismic',vmin=-0.1, vmax=0.1,aspect='auto')
ax1.set_title('VERTICAL COMPONENT')
ax1.set_ylabel('Samples, depth Sampling Rate = 1 ms')
ax1.set_xlabel('Traces, source offset = 50m')

ax2.imshow(tgdata2.T,cmap='seismic',vmin=-0.1, vmax=0.1,aspect='auto')
ax2.set_title('HORIZONTAL COMPONENT')
#ax2.set_xlabel('time (ms)')

ax3.imshow(tgdata3.T,cmap='seismic',vmin=-0.1, vmax=0.1,aspect='auto')
ax3.set_title('2ND RESIDUAL WAVEFIELD')
#ax3.set_xlabel('time (ms)')

ax4.imshow(tgdata4.T,cmap='seismic',vmin=-5, vmax=5,aspect='auto')
ax4.set_title('ZERO-PHASE UPGOING P')
#ax4.set_xlabel('time (ms)')

plt.show()
```

```python
#Showing one of each data type  
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(12,8))
fig.suptitle('EXAMPLES FOR EACH SEGY DATA TYPE')

ax1.imshow(rawdata2.T,cmap='seismic',vmin=-0.1, vmax=0.1,aspect='auto')
ax1.set_title('RAW DATA')
ax1.set_ylabel('Samples, depth Sampling Rate = 1 ms')
ax1.set_xlabel('Shot Number')

ax2.imshow(csdata1.T,cmap='seismic',vmin=-5, vmax=5,aspect='auto')
ax2.set_title('CORRIDOR STACK')
ax2.set_xlabel('Traces')

ax3.imshow(tgdata4.T,cmap='seismic',vmin=-5, vmax=5,aspect='auto')
ax3.set_title('TIME GATHER')
ax3.set_xlabel('Traces')

plt.show()
```
